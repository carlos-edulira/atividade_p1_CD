```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0,5,11) #variável x
Y = np.power(X-2,2.0) #Criando funções Y(X) e Z(X)
Z = np.exp(X/2.0)
```

```python
#criando grid
plt.plot(X,Y,label='$Y(X)=(X-2)^2$')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X)')
plt.grid(linestyle = '-' , which = 'major' , color = 'grey')
plt.legend()
plt.show()
```
