```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
# usando o seaborn para deixar melhor visualizado
import seaborn as sns
sns.set()
```

```python
#criando dataset
np.random.seed(42)

x_var = np.arange(0, 100)
y_var = x_var + np.random.normal(size = 100, loc = 0, scale = 10)

x2_var = np.arange(0,100)
y2_var = x_var + np.random.normal(size = 100, loc = 2, scale = 5)
```

```python
#plotando um scatter simples
plt.scatter(x = x_var, y = y_var)
```

```python
# dois gráficos na mesma imagen
plt.scatter(x = x_var, y = y_var)
plt.scatter(x = x2_var, y = y2_var)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.show()
```
