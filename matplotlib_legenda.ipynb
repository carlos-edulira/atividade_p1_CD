```python
import numpy as np
import matplotlib.pyplot as plt
  
X = np.linspace(0,5,11)
Y = np.power(X-2,2.0)
Z = np.exp(X/2.0)
```

* Colocando equações na legenda 

```python
plt.plot(X,Y,label='$Y(X)=(X-2)^2$')
plt.plot(X,Z,label='$Z(X)=e^{(x/2)}$')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.legend()
plt.show()
```
