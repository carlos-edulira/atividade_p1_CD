```python
#importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
#criando eixo x e y
X = np.linspace(-20,10,100)
Y = (X-3.0)*(X+2.3)
```

```python
#plotando
plt.plot(X,Y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.fill_between(X, Y)
plt.show()
```
