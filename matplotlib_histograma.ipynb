```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#números aleatórios - variável x
X = np.random.rand(5000)
```

```python
#criando gráfico
plt.hist(X)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.show()
```

```python
#alterando bins e customizando
plt.hist(X,bins=50, color='gray', edgecolor='red')
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=50')
plt.show()
```

```python
#multiplos histogramas imagens iguais
X1 = np.random.normal(0,1,size = 800)
X2 = np.random.normal(3,1,size = 800)
X3 = np.random.normal(5,1,size = 800)

plt.hist(X2,bins=10,color='yellow')
plt.hist(X3,bins=10,color='red')
plt.hist(X1,bins=10,color='purple')
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Multiplos histogramas')
plt.show()
```

```python

w = np.random.rand(1380)
plt.hist(w,bins=20,color='grey', edgecolor='orange',cumulative=True)
plt.xlabel('w')
plt.ylabel('Contagem')
plt.title('Histograma cumulativo')
plt.show()
```
