*gráfico mostrando quantidade de assaltos por mês, percentual que cada praia representa no total de assaltos


```python
import matplotlib.pyplot as plt

# Definindo lista de praias
praias = ['Tambaú', 'Bessa', 'Seixas', 'Cabo Branco', 'Jacarapé']
  
# valores da quantidade de assaltos por mês 
assaltosMes = [42, 24, 12, 30, 2]
#criando função de somar itens da lista
def somador(assaltosMes):
  soma = 0
  for i in assaltosMes:
    soma = soma + i
  return soma

soma = somador(assaltosMes)
print("Quantidade de assaltos = {0:.2f}".format(soma))
  
# cor dos item
cores = ['#DDA0DD', '#ADD8E6', '#FFFF00', '#DC143C', '#FFA500']
#afastando os itens
explode = (0.05, 0.05, 0.05, 0.05, 0.05)
  
# gráfico pizza
plt.pie(assaltosMes, colors=cores, labels=praias, autopct='%1.1f%%', pctdistance=0.85, explode=explode)
  
# desenhando circulo
circulocentral = plt.Circle((0, 0), 0.20, fc='white')
fig = plt.gcf()
  
# adicionando circulo no meio da pizza
fig.gca().add_artist(circulocentral)
  
# Titulo do gráfico
plt.title('Assaltos por Mês')
  
# Plotando gráfico
plt.show()
```
