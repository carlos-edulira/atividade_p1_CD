```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#picos de surfe e maior e menor tamanho em pés registrado em 1 ano
labels =["Chicama", "Hossegor", "Noronha", "Nazaré", "Pipeline"]
maiorPes = [5,16,14,80,15]
menorPes = [1,6,0.5,25,3]
plt.bar(labels,maiorPes)
```

```python
largura = 0.1
pos1 = np.arange(len(maiorPes))
pos2 = [x + largura for x in pos1]

plt.bar(pos1, maiorPes, color='purple', width=largura, label='Maiores Pés')
plt.bar(pos2, menorPes, color='gray', width=largura, label='Menores Pés')
plt.legend()
```

```python
plt.barh(pos1,maiorPes, label = "Barra 1")
plt.barh(pos1,menorPes, label = "Barra 2", left = maiorPes)
```
